#Transformación

1. Eliminar las últimas 3 columnas, cuyo campo son: "min…", "max…", "median…" del archivo "all_years_o3.csv" o df1

```python
df1 = df1.drop(['min (ppb)', 'max (ppb)', 'median (ppb)'], axis=1)
df1
```

2. Renombrar la columna "count" de la tabla "all_years_o3.csv" o df1 

```python
#NOTA: Es importante que las columnas coincida con los nombres de campos en la tabla merge_counts
df1 = df1.rename(columns = {'count': 'Count_o3'})
df1
```

3. Eliminar la columna "specie" de la tabla "all_years_o3.csv" o df1 

```python
df1 = df1.drop(['Specie'], axis=1)
df1
```

4. Eliminar las últimas 3 columnas, cuyo campo son: "min…", "max…", "median…" de la tabla "all_years_pm25.csv" o df2. 

```python
df2 = df2.drop(['min (ug/m3)', 'max (ug/m3)', 'median (ug/m3)'], axis=1)
df2
```

5. Renombrar la columna "count" de la tabla "all_years_pm25.csv" o df2.

```python
df2 = df2.rename(columns = {'count': 'Count_pm25'})
```

6. Eliminar la columna "specie" de la tabla‘’all_years_pm25.csv" o df2.

```python
df2 = df2.drop(['Specie'], axis=1)
df2
```

7. Unir (merge) el df1 y df2 por medio de los campos "date", "country" y "city", renombrar por new_df y mostrar en consola 

```python
new_df = df1.merge(df2, on= ['Date', 'Country', 'City'], how= 'left') #Se usó left join
new_df
```

```python
print(new_df) #Se muestra por consola el mergue entre dos dataframes
```

```python
new_df.isnull().sum() #Se verifica que no haya datos nulos
```

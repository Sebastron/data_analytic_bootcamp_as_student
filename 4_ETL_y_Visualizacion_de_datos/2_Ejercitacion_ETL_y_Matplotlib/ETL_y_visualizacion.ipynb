```python
import pandas as pd
import warnings
import plotly.express as px
warnings.simplefilter("ignore", UserWarning)
from google.colab import files
import os
#Carga de archivo por teclado
uploaded = files.upload()
for fn in uploaded.keys():
    print('User uploaded file "{name}" with length {length} bytes'.format(name=fn, 
                                                                          length=len(uploaded[fn])))
    PATH = fn
#carga y procesamiento

ext = PATH.split('.')[1]

if(ext=="xlsm"):
    df_fruit = pd.read_excel(PATH,sheet_name="Administracion y Finanzas",header=3)
    df_fruit.drop(['Unnamed: 12',2018],axis=1,inplace=True)
else:
    df_fruit = pd.read_csv(PATH, sep=";")
    df_fruit.drop(['Unnamed: 0'],axis=1,inplace=True)


df_fruit.dropna(axis=0,inplace=True)
df_fruit.rename(columns={'AÑO':'YEAR'}, inplace=True)
DIC_mes = {'Enero':1,'Febrero':2,'Marzo':3,'Abril':4,'Mayo':5,'Junio':6,'Julio':7,
           'Agosto':8,'Septiembre':9,'Octubre':10,'Noviembre':11,'Diciembre':12}
df_fruit['MES_dummy'] = df_fruit['MES'].map(DIC_mes)
df_fruit["FECHA"] = df_fruit["YEAR"].astype(int).astype(str) + "-" + df_fruit["MES_dummy"].astype(str)+"-01"
df_fruit['FECHA'] = pd.to_datetime(df_fruit['FECHA'], format='%Y/%m/%d', errors='coerce')
df_fruit.drop(['MES','MES_dummy'],axis=1,inplace=True)
df_fruit['YEAR'] = df_fruit['YEAR'].astype(int)
tortazo = pd.DataFrame(df_fruit.groupby(['YEAR']).SUELDOS.mean())
tortazo.reset_index(drop=False,inplace=True)
##Grafico 1
fig = px.line(df_fruit,x='FECHA',y=['INGRESOS','UTILIDAD'], 
              title='Ingresos y Utilidades mensuales')
fig.show()
## Grafico 2
fig = px.bar(df_fruit, x='YEAR', y='INGRESOS',title='Ventas mensuales')
fig.show()
##grafico3
fig = px.pie(tortazo, values='SUELDOS', names='YEAR',
             title='Costo de remuneraciones por año',
             color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

## Ejercicio 1

Escribir un programa que pregunte al usuario por las ventas de un rango de años y muestre por pantalla una serie con los datos de las ventas indexada por los años, antes y después de aplicarles un descuento del 10%.

```python
import pandas as pd

inicio =int(input("Ingrese el año inicial: "))
fin =int(input("Ingrese el año final: "))

ventas = {}

for i in range(inicio, fin+1):
  ventas[i] = float(input(f"Ingrese las ventas del año {i} :"))
ventas = pd.Series(ventas)

print("\nVentas sin descuento: \n", ventas)
print("\nVentas con descuento: \n", ventas*0.9)
```

## Ejercicio 2

Escribir una función que reciba una diccionario con las notas de los alumnos en curso en un examen y devuelva una serie con la nota mínima, la máxima, media y la desviación típica.

```python
import pandas as pd

def listado_notas(diccionario):
  listado = pd.DataFrame(diccionario)
  lista_salida = pd.DataFrame()
  lista_salida["Nombre Completo"] = listado["Nombre Completo"]
  #listado = listado.drop(['Nombre Completo'], axis=1) # Para eliminar columna de un DF, pero no fue necesario
  lista_salida["Nota Mínima"] = listado.min(axis = 1)
  lista_salida["Nota Maxima"] = listado.max(axis = 1) 
  lista_salida["Nota Media"] = listado.mean(axis = 1)
  lista_salida["Desviación Típica"] = listado.std(axis = 1)
  return lista_salida
parar = 0
alumnos_nombre = []
notas_1 = []
notas_2 = []
notas_3 = []
notas_4 = []
notas_5 = []
try: 
  while(parar == 0):
    nombre = input("Ingrese el nombre completo del alumno: ")
    alumnos_nombre.append(nombre)
    nota_1 = abs(float(input("Ingrese la nota N°1 del alumno (use los puntos para decimales): ")))
    notas_1.append(nota_1)
    nota_2 = abs(float(input("Ingrese la nota N°2 del alumno (use los puntos para decimales): ")))
    notas_2.append(nota_2)
    nota_3 = abs(float(input("Ingrese la nota N°3 del alumno (use los puntos para decimales): ")))
    notas_3.append(nota_3)
    nota_4 = abs(float(input("Ingrese la nota N°4 del alumno (use los puntos para decimales): ")))
    notas_4.append(nota_4)
    nota_5 = abs(float(input("Ingrese la nota N°5 del alumno (use los puntos para decimales): ")))
    notas_5.append(nota_5)
    continuar = input("¿Deseas agregar más alumno? (si/no):").upper()
    if(continuar == "NO"):
      parar = 1
    d = {'Nombre Completo' : alumnos_nombre, "Nota N°1":notas_1, "Nota N°2":notas_2,  "Nota N°3":notas_3,  "Nota N°4":notas_4, "Nota N°5":notas_5}
  lista = listado_notas(d)
  print("\n")
  print(lista)
  print("\n")
except ValueError:
  print("Error en el ingreso de las notas, debe ser entero o decimal, vuelve a intentar.") 
```

## Ejercicio 3

Escribir una función que reciba una diccionario con las notas de los alumnos en curso en un examen y devuelva una serie con las notas de los alumnos aprobados ordenadas de mayor a menor.

```python
import pandas as pd

def listado_notas(diccionario):
  listado = pd.DataFrame(diccionario)
  lista_salida = pd.DataFrame()
  listado["Nota Media"] = listado.mean(axis = 1)
  listado = listado.drop(listado[listado['Nota Media'] < 4].index)
  lista_salida["Nombre Completo"] = listado["Nombre Completo"]
  lista_salida["Nota Promedio"] = listado['Nota Media']
  lista_salida = lista_salida.sort_values('Nota Promedio',ascending=False)
  return lista_salida
parar = 0
alumnos_nombre = []
notas_1 = []
notas_2 = []
notas_3 = []
notas_4 = []
notas_5 = []
try: 
  while(parar == 0):
    nombre = input("Ingrese el nombre completo del alumno: ")
    alumnos_nombre.append(nombre)
    nota_1 = abs(float(input("Ingrese la nota N°1 del alumno (use los puntos para decimales): ")))
    notas_1.append(nota_1)
    nota_2 = abs(float(input("Ingrese la nota N°2 del alumno (use los puntos para decimales): ")))
    notas_2.append(nota_2)
    nota_3 = abs(float(input("Ingrese la nota N°3 del alumno (use los puntos para decimales): ")))
    notas_3.append(nota_3)
    nota_4 = abs(float(input("Ingrese la nota N°4 del alumno (use los puntos para decimales): ")))
    notas_4.append(nota_4)
    nota_5 = abs(float(input("Ingrese la nota N°5 del alumno (use los puntos para decimales): ")))
    notas_5.append(nota_5)
    continuar = input("¿Deseas agregar más alumno? (si/no):").upper()
    if(continuar == "NO"):
      parar = 1
    d = {'Nombre Completo' : alumnos_nombre, "Nota N°1":notas_1, "Nota N°2":notas_2,  "Nota N°3":notas_3,  "Nota N°4":notas_4, "Nota N°5":notas_5}
  lista = listado_notas(d)
  print("\nListado de alumnos que aprobaron en la asignatura.\n")
  print(lista)
  print("\n")
except ValueError:
  print("Error en el ingreso de las notas, debe ser entero o decimal, vuelve a intentar.") 
```

## Ejercicio 5

Escribir una función que reciba un DataFrame con el formato del ejercicio anterior, una lista de meses, y devuelva el balance (ventas - gastos) total en los meses indicados.

```python
import pandas as pd
def balance(df):
  df["Balance"] = df["Ventas"] - df["Gastos"] 
  return df

lista1 = [["Enero", 30500, 22000], ["Febrero", 35600, 23400], ["Marzo", 28300, 18100], ["Abril", 33900, 20700]]
lista2 = ["Mes", "Ventas", "Gastos"]
data_i = pd.DataFrame(lista1, columns = lista2)
data = balance(data_i)
data
```

El archivo <a href="https://drive.google.com/file/d/1V06-QPe36PIIMq0z6Nm8_02mHB1KX_Cu/view?usp=sharing"> cotizacion.csv</a> contiene las cotizaciones de las empresas del IBEX35 con las siguientes columnas: <font color=#B40431>Nombre</font> (nombre de la empresa), <font color=#B40431>Final</font> (precio de la acción al cierre de bolsa), <font color=#B40431>Máximo</font> (precio máximo de la acción durante la jornada), <font color=#B40431>Mínimo</font> (precio mínimo de la acción durante la jornada), <font color=#B40431>Volumen</font> (Volumen al cierre de bolsa), <font color=#B40431>Efectivo</font> (capitalización al cierre en miles de euros). Construir una función que construya un DataFrame a partir del un archivo con el formato anterior y devuelva otro DataFrame con el mínimo, el máximo y la media de dada columna.

```python
import pandas as pd

def estadisticas(cotizacion):
  cotizacion = pd.read_csv(cotizacion, sep=';', decimal=',', thousands='.', index_col=0)
  #print(cotizacion)
  estadisticas = pd.DataFrame([cotizacion.min(), cotizacion.max(), cotizacion.mean()], index = ['Minimo', 'Maximo', 'Media'])

  return estadisticas

estadisticas('cotizacion.csv')
```

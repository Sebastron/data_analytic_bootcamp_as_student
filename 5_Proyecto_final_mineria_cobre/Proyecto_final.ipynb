##Transformación de datos

##Archivo 1: Producción Nacional de cobre por mineras

Mergue entre DF6, DF7, DF8 Y DF9

```python
produccion = df6.merge(df7, on= ['Mes-año'], how= 'inner')
produccion = produccion.merge(df8, on= ['Mes-año'], how= 'inner')
produccion = produccion.merge(df9, on= ['Mes-año'], how= 'inner')
produccion
```

```python
produccion["Mes-año"] = produccion["Mes-año"].str.split("/").str.get(2).astype(int)
produccion = produccion.rename(columns={'Mes-año':'Año'})
produccion = produccion[(produccion.Año >= 2010) & (produccion.Año <=2021)]
produccion
```

```python
columnas = produccion.Año.unique()
aux_year = produccion.Año.tolist()
#print(columnas)
for i in columnas:
  produccion[(produccion['Año'] == i)] = produccion[produccion["Año"] == i].mean()
produccion["Año"] = aux_year
produccion = produccion.drop_duplicates()
produccion
```

```python
produccion.to_csv("Produccion Nacional Generales 2010 - 2021.csv", index = False)
```

```python
produccion.to_excel("Produccion Nacional Generales 2010 - 2021.xls", index = False)
```

##Archivo 2: Exportación de cobre

Transformando DF4

```python
df4["Serie"] = df4["Serie"].str.strip(" (1)").str.strip(" (2)").str.strip(" (3)")
df4["Continente"] = df4.Serie.str.split(" - ").str.get(0)
df4["Pais"] = df4.Serie.str.split(" - ").str.get(1)
df4 = df4.drop(["Serie"], axis=1)
df4
```

```python
paises_unicos = df4.Pais.unique()
aux_continente = df4.Continente.tolist()
aux_pais = df4.Pais.tolist()
for i in paises_unicos:
  df4[(df4['Pais'] == i)] = df4[df4["Pais"] == i].sum()
df4["Pais"] = aux_pais
df4["Continente"] = aux_continente
df4 = df4.drop_duplicates()
df4
```

```python
columnas= df4.columns.tolist()
column = columnas[12:]
#print(column)
exportaciones_valor = pd.melt(df4, id_vars=column, var_name='Año', value_name='Valor (Millones $USD)')
exportaciones_valor = exportaciones_valor.reindex(columns=['Año','Continente','Pais','Valor (Millones $USD)'])
exportaciones_valor
```

Transformando DF5

```python
df5["Serie"] = df5["Serie"].str.strip(" (1)").str.strip(" (2)").str.strip(" (3)")
df5["Continente"] = df5.Serie.str.split(" - ").str.get(0)
df5["Pais"] = df5.Serie.str.split(" - ").str.get(1)
df5 = df5.drop(["Serie"], axis=1)
df5
```

```python
paises_unicos = df5.Pais.unique()
aux_continente = df5.Continente.tolist()
aux_pais = df5.Pais.tolist()
for i in paises_unicos:
  df5[(df5['Pais'] == i)] = df5[df5["Pais"] == i].sum()
df5["Pais"] = aux_pais
df5["Continente"] = aux_continente
df5 = df5.drop_duplicates()
df5
```

```python
columnas= df5.columns.tolist()
column = columnas[12:]
#print(column)
exportaciones_cantidad= pd.melt(df5, id_vars=column, var_name='Año', value_name='Cantidad (Miles TM)')
exportaciones_cantidad = exportaciones_cantidad.reindex(columns=['Año','Continente','Pais','Cantidad (Miles TM)'])
exportaciones_cantidad
```

```python
exportaciones = exportaciones_cantidad.merge(exportaciones_valor[['Año','Continente', 'Pais', 'Valor (Millones $USD)']], how= 'inner')
exportaciones
```

```python
exportaciones.to_excel("Exportaciones Generales 2010 - 2021.xls", index = False)
```

```python
exportaciones.to_csv("Exportaciones Generales 2010 - 2021.csv", index = False)
```

###Archivo 3: Datos Globales

```python
df11 = df11.drop(["Total Codelco", "Total privadas"], axis=1)
df11 = df11.rename(columns = {"Total país": "Produccion Nacional (Miles TM)"})
df11
```

```python
Totales_exportacion = df1[df1.Serie == "Totales"]
columnas= Totales_exportacion.columns.tolist()
column = columnas[:1]
#print(column)
Totales_exportacion = pd.melt(Totales_exportacion, id_vars=column, var_name='Año', value_name='Cobre Exportado (Miles TM)')
Totales_exportacion = Totales_exportacion.drop(["Serie"], axis=1)
Totales_exportacion["Año"] = Totales_exportacion["Año"].astype(int)
Totales_exportacion = Totales_exportacion[(Totales_exportacion.Año >= 2010) & (Totales_exportacion.Año <=2021)]
Totales_exportacion
```

```python
Totales_ingresos = df2[df2.Serie == "Totales"]
columnas= Totales_ingresos.columns.tolist()
column = columnas[:1]
#print(column)
Totales_ingresos = pd.melt(Totales_ingresos, id_vars=column, var_name='Año', value_name='Ingresos exportación (Millones de dolares)')
Totales_ingresos = Totales_ingresos.drop(["Serie"], axis=1)
Totales_ingresos["Año"] = Totales_ingresos["Año"].astype(int)
Totales_ingresos = Totales_ingresos[(Totales_ingresos.Año >= 2010) & (Totales_ingresos.Año <=2021)]
Totales_ingresos
```

```python
Precio_cobre = df3
Precio_cobre = Precio_cobre[(Precio_cobre.AÑO >= 2010) & (Precio_cobre.AÑO <=2021)]
Precio_cobre = Precio_cobre.rename(columns={'AÑO':'Año','B.M.L. / LME':'Precio Cobre LME','COMEX':'Precio Cobre COMEX'})
Precio_cobre
```

```python
exportacion_continente = exportaciones.groupby(['Año','Continente'], sort = False)['Cantidad (Miles TM)','Valor (Millones $USD)'].sum()
exportacion_continente = exportacion_continente.reset_index(drop=False)
exportacion_continente
```

```python
x = exportacion_continente.pivot(index = 'Año', columns='Continente', values=["Cantidad (Miles TM)",	"Valor (Millones $USD)"]).reset_index(drop=False)
```

```python
aux = x.to_numpy()
aux.shape
```

```python
exportacion_continente = pd.DataFrame(aux, columns=['Año', 'Exportación América (Miles TM)', 'Exportación Asia (Miles TM)', 'Exportación Europa (Miles TM)', 'Ingresos América (Millones $USD)', 'Ingresos Asia (Millones $USD)', 'Ingresos Europa (Millones $USD)'])
exportacion_continente
```

```python
produccion_mundial = df10[(df10["PRODUCCION MUNDIAL Y CHILENA DE COBRE DE MINA (porcentaje y tonelaje)"] == "PRODUCCION MUNDIAL")]
produccion_mundial = produccion_mundial.rename(columns={'PRODUCCION MUNDIAL Y CHILENA DE COBRE DE MINA (porcentaje y tonelaje)':'Tipo'})
produccion_mundial = produccion_mundial.drop(["Unidad", "Tipo"], axis=1)
produccion_mundial = pd.melt(produccion_mundial, var_name='Año', value_name='Produccion Mundial (Miles TM)')
produccion_mundial = produccion_mundial[(produccion_mundial["Año"]>=2010) & (produccion_mundial["Año"]<=2020)]
produccion_mundial
```

```python
datos_globales = Totales_exportacion.merge(Totales_ingresos, on = "Año", how= 'inner')
datos_globales = datos_globales.merge(df11, on="Año", how="inner")
datos_globales = Precio_cobre.merge(datos_globales, on="Año", how="inner")
datos_globales["Año"] = datos_globales["Año"].astype(int)
exportacion_continente["Año"] = exportacion_continente["Año"].astype(int)
datos_globales = datos_globales.merge(exportacion_continente, on="Año", how="inner")
datos_globales = datos_globales.merge(produccion_mundial, on="Año", how="left").sort_values(["Año"], ascending=True).fillna(0).reset_index(drop=True)
datos_globales['Produccion Nacional (Miles TM)'] = datos_globales['Produccion Nacional (Miles TM)'].apply(lambda x: (x.split()[0].replace('.','')))
datos_globales['Produccion Nacional (Miles TM)'] = datos_globales['Produccion Nacional (Miles TM)'].apply(lambda x: float(x.split()[0].replace(',','.')))
datos_globales
```

```python
datos_globales = datos_globales.reindex(columns=['Año','Produccion Nacional (Miles TM)','Produccion Mundial (Miles TM)','Cobre Exportado (Miles TM)','Ingresos exportación (Millones de dolares)','Exportación América (Miles TM)', 'Exportación Asia (Miles TM)', 'Exportación Europa (Miles TM)', 'Ingresos América (Millones $USD)', 'Ingresos Asia (Millones $USD)', 'Ingresos Europa (Millones $USD)', 'Precio Cobre LME', 'Precio Cobre COMEX'])
datos_globales
```

```python
datos_globales.to_csv("Tabla Principal.csv", sep=";", index=False)
```

```python
datos_globales.to_excel("Tabla Principal.xls", index=False)
```

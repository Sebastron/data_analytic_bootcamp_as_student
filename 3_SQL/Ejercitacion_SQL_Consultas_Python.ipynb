21. ¿Cuántos clientes hay de Argentina, Brazil, Bolivia o Chile? ¿Hay algún otro país de latinoamérica?

```python
#Para el caso de Argentina, Brazil, Bolivia y Chile
consulta21_1 = pd.read_sql("select country as País, count(name) as 'Clientes por País' from customer_list where country in ('Argentina','Bolivia','Brazil','Chile') group by country;", conexion)
consulta21_1
```

```python
#Para buscar otros paises latinoamericanos
consulta21_2 = pd.read_sql("SELECT DISTINCT country FROM country;", conexion)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
print(consulta21_2)
pd.set_option('display.max_rows', 20)
```

Como resultado en esta exploración de datos, si existe registros de otros paises de latinoamérica, ya que se encontraron ‘Venezuela’, ‘Puerto Rico’, ‘Peru’, ‘Paraguay’, ‘Mexico’, ‘Colombia’, ‘Ecuador’ y ‘Dominican Republic’.
Por lo tanto, para todos los paises latinoamericanos, incluyendo los que se han encontrados, la cantidad de clientes se presenta de la siguiene manera

```python
consulta21_3 = pd.read_sql("select country as País, count(name) as 'Clientes por País' from customer_list where country in ('Argentina','Bolivia', 'Brazil', 'Chile', 'Venezuela', 'Puerto Rico', 'Peru', 'Paraguay', 'Mexico', 'Colombia', 'Ecuador', 'Dominican Republic') group by country;", conexion)
consulta21_3
```
